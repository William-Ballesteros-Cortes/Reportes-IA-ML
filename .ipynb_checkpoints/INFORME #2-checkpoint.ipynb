```python
import sklearn
import numpy as np
import pandas as pd
```

Se carga el archivo y hacemos la separación de datos de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
dataALL=pd.read_csv("https://github.com/William-Ballesteros-Cortes/""Reportes-IA-ML/blob/informe1/games.csv?raw=true", sep = ";")  
N=len(dataALL)
cTrain=int(N*0.8) # 80% para entrenar y 20% para probar
cTest=N-cTrain
print(N,cTrain,cTest)
train_data,test_data= sklearn.model_selection.train_test_split(dataALL, train_size=cTrain, test_size=cTest)
```

## Definimos las pipeline para los atributos numéricos

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#11 Atributos Numericos
num_attributes = ['gameDuration',  
                  't1_towerKills',
                  't1_inhibitorKills', 
                  't1_baronKills', 
                  't1_dragonKills', 
                  't1_riftHeraldKills', 
                  't2_towerKills', 
                  't2_inhibitorKills', 
                  't2_baronKills', 
                  't2_dragonKills', 
                  't2_riftHeraldKills']

num_pipeline= Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()) 
    
    ])
```

## Definimos las pipeline para los atributos categóricos

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# 6 Atributos Categóricos
cat_attributes = ['firstBlood',
                  'firstTower', 
                  'firstInhibitor', 
                  'firstBaron',
                  'firstDragon', 
                  'firstRiftHerald']

cat_pipeline= Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("cat_encoder", OneHotEncoder(sparse=False))
    ])
```

### Pipeline Completo

```python
from sklearn.compose import ColumnTransformer

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attributes),
    ("cat", cat_pipeline, cat_attributes), 
])
```

## Se extraen las etiquetas de clase

```python
X_train = full_pipeline.fit_transform(train_data)
```

```python
X_train.shape
```

```python
X_train[0,:]
```

```python
y_train = train_data["winner"]
y_train
```
